## <font color='blue'>Redox Flow Batteries (RFBs)

RFBs work similarily to standard electrochemical batteries. They rely on a redox reaction for electron flow. You can improve the storage capacity of a RFB if you use a redox pair with a higher redox potential.

## <font color='blue'>Three types of electrolytes 

Aqueous : Low energy denisty and limited by electrochemimcal window of water.

Non-Aqueous : Better voltage, but low solubility of active materials and can be temperature dependent.

Organic (type of non-aqueous) : High volatage but low effiency. Expensive and volatile.

### <font color='red'>Ionic liquids are a promising class of non-aqueous electrolytes. They are poorly coordinated salts.

Non-volitile and non-flammable.

Stable, high density, and high electrical conductivity.

Broad electrochemical window (prevents degradation of the battery).

But complex synthesis and possible toxicity.
